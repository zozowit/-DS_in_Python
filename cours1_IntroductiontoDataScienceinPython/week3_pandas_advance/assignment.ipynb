```python
import pandas as pd
import numpy as np

energy = pd.read_excel("Energy Indicators.xls")
energy = energy.drop('Unnamed: 0', axis=1)
energy = energy.drop('Unnamed: 1', axis=1)

energy.columns=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
# make sure to exclude the footer and header information from the datafile.
# energy.drop?

energy = energy.iloc[16:]

# reset the index start from 0
energy.reset_index(drop=True, inplace=True)

# Convert `Energy Supply` to gigajoules (there are 1,000,000 gigajoules in a petajoule).
# For all countries which have missing data (e.g. data with "...") 
# make sure this is reflected as `np.NaN` values.
energy['Energy Supply'] = energy['Energy Supply'].map(lambda x: np.NaN if x=='...' else x/1000000)
energy[energy['Country'] == 'United Kingdom of Great Britain and Northern Ireland19']

# find the country  has number
change_mask = energy['Country'].str.contains(r'[0-9]+| [(,))]+', na=False)
energy[change_mask]

# here are also several countries with numbers and/or parenthesis in their name. Be sure to remove these
# 'Bolivia (Plurinational State of)' should be 'Bolivia',

# 'Switzerland17' should be 'Switzerland'.
energy['Country'] = energy['Country'].str.replace('[0-9]+|[(,)]+', '')
energy[change_mask]

# Rename the following list of countries (for use in later questions):
energy.replace({"Republic of Korea": "South Korea", 
                "United States of America": "United States",
                "United Kingdom of Great Britain and Northern Ireland": "United Kingdom", 
                "China Hong Kong Special Administrative Region": "Hong Kong",
                "Iran Islamic Republic of": 'Iran'}, 
               inplace=True)

# check all have been changed
country_mask = energy['Country'].isin(['South Korea', "United States","United Kingdom", "Hong Kong"])
energy[country_mask]
# energy[energy['Country'] == 'Hong Kong']
#energy['Country'].iloc[235]

# load the GDP data from the file world_bank.csv,
# which is a csv containing countries' GDP from 1960 to 2015 from World Bank. 
# Call this DataFrame GDP
GDP = pd.read_csv('world_bank.csv')
# GDP.head()
GDP_country_list = list(GDP.iloc[3])
print(GDP_country_list)
GDP = GDP.iloc[4:]
#print(GDP.iloc[0])
GDP.columns = GDP_country_list
#print(GDP.iloc[0])

# rename these country
# "Korea, Rep.": "South Korea", "Iran, Islamic Rep.": "Iran", "Hong Kong SAR, China": "Hong Kong"
GDP['Country Name'].replace({"Korea, Rep.": "South Korea", 
                                                "Iran, Islamic Rep.": "Iran", 
                                                "Hong Kong SAR, China": "Hong Kong"}, 
                                              inplace=True)
# check all have been changed, test code
# GDP_country_mask = GDP['Country Name'].isin(['South Korea', "Iran", "Hong Kong"])
# GDP[GDP_country_mask]

# load the Sciamgo from the file scimagojr-3.xlsx,
# which ranks countries based on their journal contributions in the aforementioned area. 
# Call this DataFrame ScimEn.
ScimEn = pd.read_excel('scimagojr-3.xlsx')
ScimEn.head()
```

```python
# get the top 15 country in ScimEn rank
rank_mask = ScimEn['Rank'] < 16
rank_country_name = list(ScimEn[rank_mask]['Country'])
print(rank_country_name)
ScimEn_rank = ScimEn[rank_mask]
ScimEn_rank.reset_index(drop=True, inplace=True)

energy_rank = energy[energy['Country'].isin(rank_country_name)]
energy_rank.reset_index(drop=True, inplace=True)

GDP_rank = GDP[GDP['Country Name'].isin(rank_country_name)]
GDP_rank.reset_index(drop=True, inplace=True)

# set GDP_rank column name to 'Country'
GDP_rank = GDP_rank.rename(columns={'Country Name': 'Country'})
print('GDP_rank columns: ' + str(list(GDP_rank.columns)))
# energy_rank
# find the missing country in energy is 'Iran Islamic Republic of'
# go backk to energy to rename it to Iran
#energy[energy['Country'].str.contains('Ir', na=False)]
#GDP_rank
#check
# ScimEn_rank
print(list(ScimEn_rank.columns))
```

```python
all_rank = pd.merge(energy_rank, ScimEn_rank, how='right', left_on='Country', right_on='Country')
print(list(all_rank.columns))
all_rank = pd.merge(GDP_rank, all_rank, how='right', left_on='Country', right_on='Country')
print(list(all_rank.columns))
#all_rank

# The index of this DataFrame should be the name of the country, 
all_rank.set_index('Country', inplace=True)

# all_rank.set_index('')

# 2006.0, 2007.0, 2008.0, 2009.0, 2010.0, 2011.0, 2012.0, 2013.0, 2014.0, 2015.0
# they are <class 'numpy.float64'> which should make them to interger
for col in all_rank.columns:
    # print(str(type(col)) + '-\n')
    if type(col) == np.float64:
        all_rank.rename(columns={col: col.astype(int)}, inplace=True)
        #print('float64 to'+ str(type(col)))

print(list('change to int\n'))
print(list(all_rank.columns))

# the columns should be ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 
# 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', 
#'2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015'].
all_rank_columns = ['Rank', 'Documents', 'Citable documents', 
                    'Citations', 'Self-citations', 'Citations per document', 
                    'H index', 'Energy Supply', 'Energy Supply per Capita', 
                    '% Renewable', 2006, 2007, 
                    2008, 2009, 2010, 
                    2011, 2012, 2013, 
                    2014, 2015]
# note that the column about year is type int, so the all_rank_columns list should be int
all_rank = all_rank[all_rank_columns]
all_rank
```
