```python
import pandas as pd
import numpy as np

energy = pd.read_excel("Energy Indicators.xls")
energy = energy.drop('Unnamed: 0', axis=1)
energy = energy.drop('Unnamed: 1', axis=1)

energy.columns=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
# make sure to exclude the footer and header information from the datafile.
# energy.drop?

energy = energy.iloc[16:]

# reset the index start from 0
energy.reset_index(drop=True, inplace=True)

# Convert `Energy Supply` to gigajoules (there are 1,000,000 gigajoules in a petajoule).
# For all countries which have missing data (e.g. data with "...") 
# make sure this is reflected as `np.NaN` values.
energy['Energy Supply'] = energy['Energy Supply'].map(lambda x: np.NaN if x=='...' else x*1000000)
energy[energy['Country'] == 'United Kingdom of Great Britain and Northern Ireland19']

# find the country  has number
change_mask = energy['Country'].str.contains(r'[0-9]+| [(,))]+', na=False)
energy[change_mask]

# here are also several countries with numbers and/or parenthesis in their name. Be sure to remove these
# 'Bolivia (Plurinational State of)' should be 'Bolivia',

# 'Switzerland17' should be 'Switzerland'.
energy['Country'] = energy['Country'].str.replace('[0-9]+|[(,)]+', '')
energy[change_mask]

# Rename the following list of countries (for use in later questions):
energy.replace({"Republic of Korea": "South Korea", 
                "United States of America": "United States",
                "United Kingdom of Great Britain and Northern Ireland": "United Kingdom", 
                "China Hong Kong Special Administrative Region": "Hong Kong",
                "Iran Islamic Republic of": 'Iran'}, 
               inplace=True)

# check all have been changed
country_mask = energy['Country'].isin(['South Korea', "United States","United Kingdom", "Hong Kong"])
energy[country_mask]
# energy[energy['Country'] == 'Hong Kong']
#energy['Country'].iloc[235]

# load the GDP data from the file world_bank.csv,
# which is a csv containing countries' GDP from 1960 to 2015 from World Bank. 
# Call this DataFrame GDP
GDP = pd.read_csv('world_bank.csv')
# GDP.head()
GDP_country_list = list(GDP.iloc[3])
print(GDP_country_list)
GDP = GDP.iloc[4:]
#print(GDP.iloc[0])
GDP.columns = GDP_country_list
#print(GDP.iloc[0])

# rename these country
# "Korea, Rep.": "South Korea", "Iran, Islamic Rep.": "Iran", "Hong Kong SAR, China": "Hong Kong"
GDP['Country Name'].replace({"Korea, Rep.": "South Korea", 
                                                "Iran, Islamic Rep.": "Iran", 
                                                "Hong Kong SAR, China": "Hong Kong"}, 
                                              inplace=True)
# check all have been changed, test code
# GDP_country_mask = GDP['Country Name'].isin(['South Korea', "Iran", "Hong Kong"])
# GDP[GDP_country_mask]

# load the Sciamgo from the file scimagojr-3.xlsx,
# which ranks countries based on their journal contributions in the aforementioned area. 
# Call this DataFrame ScimEn.
ScimEn = pd.read_excel('scimagojr-3.xlsx')
ScimEn.head()
# get the top 15 country in ScimEn rank
rank_mask = ScimEn['Rank'] < 16
rank_country_name = list(ScimEn[rank_mask]['Country'])
print(rank_country_name)
ScimEn_rank = ScimEn[rank_mask]
ScimEn_rank.reset_index(drop=True, inplace=True)

energy_rank = energy[energy['Country'].isin(rank_country_name)]
energy_rank.reset_index(drop=True, inplace=True)

GDP_rank = GDP[GDP['Country Name'].isin(rank_country_name)]
GDP_rank.reset_index(drop=True, inplace=True)

# set GDP_rank column name to 'Country'
GDP_rank = GDP_rank.rename(columns={'Country Name': 'Country'})
print('GDP_rank columns: ' + str(list(GDP_rank.columns)))
# energy_rank
# find the missing country in energy is 'Iran Islamic Republic of'
# go backk to energy to rename it to Iran
#energy[energy['Country'].str.contains('Ir', na=False)]
#GDP_rank
#check
# ScimEn_rank
print(list(ScimEn_rank.columns))

all_rank = pd.merge(energy_rank, ScimEn_rank, how='right', left_on='Country', right_on='Country')
print(list(all_rank.columns))
all_rank = pd.merge(GDP_rank, all_rank, how='right', left_on='Country', right_on='Country')
print(list(all_rank.columns))
#all_rank

# The index of this DataFrame should be the name of the country, 
all_rank.set_index('Country', inplace=True)

# all_rank.set_index('')

# 2006.0, 2007.0, 2008.0, 2009.0, 2010.0, 2011.0, 2012.0, 2013.0, 2014.0, 2015.0
# they are <class 'numpy.float64'> which should make them to interger
for col in all_rank.columns:
    # print(str(type(col)) + '-\n')
    if type(col) == np.float64:
        all_rank.rename(columns={col: col.astype(int)}, inplace=True)
        #print('float64 to'+ str(type(col)))

print(list('change to int\n'))
print(list(all_rank.columns))

# the columns should be ['Rank', 'Documents', 'Citable documents', 'Citations', 'Self-citations', 
# 'Citations per document', 'H index', 'Energy Supply', 'Energy Supply per Capita', '% Renewable', 
#'2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015'].
all_rank_columns = ['Rank', 'Documents', 'Citable documents', 
                    'Citations', 'Self-citations', 'Citations per document', 
                    'H index', 'Energy Supply', 'Energy Supply per Capita', 
                    '% Renewable', 2006, 2007, 
                    2008, 2009, 2010, 
                    2011, 2012, 2013, 
                    2014, 2015]
# note that the column about year is type int, so the all_rank_columns list should be int
all_rank15 = all_rank[all_rank_columns]


def answer_one():
    return all_rank15

answer_one()
```

```python
# What is the average GDP over the last 10 years for each country? 
# (exclude missing values from this calculation.)
# heir average GDP sorted in descending order.
all_rank_years_columns = [2006, 2007, 
                    2008, 2009, 2010, 
                    2011, 2012, 2013, 
                    2014, 2015]
def answer_three():
    Top15 = answer_one()
    Top15['average GDP'] = Top15[all_rank_years_columns].mean(1)
    Top15.sort_index(by='average GDP', ascending=False, inplace=True)
    return Top15
answer_three()
```

```python
# By how much had the GDP changed over the 10 year span 
# for the country with the 6th largest average GDP?
def answer_four():
    Top15 = answer_three()
    Top6 = Top15.iloc[:5]
    Top6['GDP_span'] = Top6[2015] - Top6[2006]
    return Top6
answer_four()
```

```python
#What is the mean Energy Supply per Capita?
def answer_five():
    Top15 = answer_one()
    return Top15['Energy Supply per Capita'].mean()

answer_five()
```

```python
# What country has the maximum % Renewable and what is the percentage?

# This function should return a tuple with the name of the country and the percentage.
def answer_six():
    Top15 = answer_one()
    max_num = Top15['% Renewable'].max()
    Country_num = Top15[Top15['% Renewable'] == max_num].index[0]
    return (Country_num, max_num)

answer_six()
```

```python
# Question 7 (6.6%)
# Create a new column that is the ratio of Self-Citations to Total Citations. 
# What is the maximum value for this new column, and what country has the highest ratio?
# This function should return a tuple with the name of the country and the ratio.
def answer_seven():
    Top15 = answer_one()
    print(list(Top15.columns))
    Top15['ratio'] = Top15['Self-citations'] / Top15['Citations']
    max_ratio = Top15['ratio'].max()
    print(max_ratio)
    max_s = Top15[Top15['ratio'] == max_ratio]
    return (max_s.index[0], max_ratio)

answer_seven()
```

```python
# Question 8 (6.6%)
# Create a column that estimates the population using Energy Supply 
# and Energy Supply per capita. What is the third most populous country according to this estimate?

# This function should return a single string value.
def answer_eight():
    Top15 = answer_one()
    # print(Top15.columns)
    Top15['population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    
    Top15.sort_index(by='population', inplace=True, ascending=False)
    #print(Top15.head())
    #print(type(Top15.iloc[2]))
    
    # Top15.iloc[2] would return all the velue in index 2 but not the object of serie 2
    return Top15.index.tolist()[2]

answer_eight()
```

```python
# Question 9 (6.6%)¶
# Create a column that estimates the number of citable documents per person. 
# What is the correlation between the number of citable documents per capita and 
# the energy supply per capita? Use the .corr() method, (Pearson's correlation).

# This function should return a single number.

# (Optional: Use the built-in function plot9() to visualize the relationship 
# between Energy Supply per Capita vs. Citable docs per Capita)
def answer_nine():
    Top15 = answer_one()
    #print(list(Top15.columns))
    Top15['population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['citable documents per capita'] = Top15['Citable documents'] / Top15['population'] 
    # print(Top15.head())
    #print(list(Top15.columns))
    corr_df = Top15[['Energy Supply per Capita', 'citable documents per capita']]
    #corr_df = Top15[['Energy Supply per Capita', 'Citable documents']]
    print(corr_df.head())
    print(corr_df.corr())
    df = corr_df.corr()
    return df

answer_nine().head()
```

```python
# Create a new column with a 1 if the country's % Renewable value is at or above the median 
# for all countries in the top 15, and a 0 if the country's % Renewable value is below the median.

# This function should return a series named HighRenew 
# whose index is the country name sorted in ascending order of rank.
def answer_ten():
    Top15 = answer_one()
#    print(list(Top15.columns))
    median = Top15['% Renewable'].median()
    Top15['HighRenew'] = Top15['% Renewable'].map(lambda x: 1 if (x > median) else 0)
    Top15.sort_index(by='Rank', ascending=True, inplace=True)
#    print(Top15)
    return Top15['HighRenew']

answer_ten()
```

```python
# Question 13 (6.6%)
# Convert the Population Estimate series to a string with thousands separator (using commas). Do not round the results.

# e.g. 317615384.61538464 -> 317,615,384.61538464

# This function should return a Series PopEst whose index is the country name and whose values are the population estimate string.

def thousands_separator(x):
    print("%d / 1000 = %d"%(x, x/1000))
    str_x = str(x)
    return str(x)
    

def answer_thirteen():
    Top15 = answer_one()
    Top15['est_pop'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']

    # ！！！！！！！！！！！！！！！！！！！！！！！
    # 'est_pop' is <class 'float'> which could not be average in agg
    # 'Energy Supply' is <class 'numpy.float64'>
    Top15['est_pop'] = Top15['est_pop'].astype('float64')
    print(type(Top15['est_pop']))
    Top15['est_pop_str'] = Top15['est_pop'].map(thousands_separator)
    
    return  Top15['est_pop_str']

est_pop = answer_thirteen()

print(type(est_pop[0]))
est_pop.head()
```
